## Building the different models

### Model 1: Linear regression

```python
train_time = [i+1 for i in range(len(train))]
test_time = [i+133 for i in range(len(test))]
print('Training time instance','\n', train_time,'\n')
print('Test time instance', '\n', test_time, '\n')
```

```python
lr_train = train.copy()
lr_test = test.copy()
```

```python
lr_train['time'] = train_time
lr_test['time'] = test_time

print('First few rows of training data','\n', lr_train.head(),'\n')
print('Last few rows of training data', '\n', lr_train.tail(), '\n')
print('First few rows of test data', '\n', lr_test.head(), '\n')
print('Last few rows of test data', '\n', lr_test.tail(), '\n')
```

```python
lr_train.head()
```

```python
lr_test.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(lr_train[['time']], lr_train['Rose'].values)
```

```python
lr_train[['time']].head()
```

```python
train_pred_model = lr.predict(lr_train[['time']])
lr_train['RegOnTime'] = train_pred_model

test_pred_model = lr.predict(lr_test[['time']])
lr_test['RegOnTime'] = test_pred_model

plt.plot(train['Rose'], label = 'Train')
plt.plot(test['Rose'], label = 'Test')
plt.plot(lr_test['RegOnTime'], label='Regression On Time_Test Data')
# Plotting the predicted test data
plt.plot(lr_train['RegOnTime'], label='Regression On Time_Training Data')
# Setting the location of the legend
plt.legend(loc='best')
plt.grid();
```

### Calculating accuracy metrics of linear regression model

```python
from sklearn import metrics
```

```python
## Calculating MAPE
def calculate_mape(y_true, y_pred):
    return np.mean((np.abs(y_true- y_pred))/(y_true))*100
```

```python
### Model evaluation on train data using RMSE and MAPE
rmse_model1_train = metrics.mean_squared_error(train['Rose'], train_pred_model, squared=False)
mape_model1_train = calculate_mape(train['Rose'], train_pred_model)
print("For RegressionOnTime forecast on the training data, RMSE is %3.3f MAPE is %3.2f"%(rmse_model1_train, mape_model1_train))
```

```python
### Model evaluation on test data using RMSE and MAPE
rmse_model1_test = metrics.mean_squared_error(test['Rose'], test_pred_model, squared=False)
mape_model1_test = calculate_mape(test['Rose'], test_pred_model)
print("For RegressionOnTime forecast on the testing data, RMSE is %3.3f MAPE is %3.2f"%(rmse_model1_test, mape_model1_test))
```

```python
RMSE=((sum((test['Rose']-test_pred_model)**2))/len(test_pred_model))**(0.5)
print(RMSE)
```

```python
lrResultsDf = pd.DataFrame({'Test RMSE': [rmse_model1_test], 'Test MAPE':[mape_model1_test]}, index=['RegressionOnTime'])
lrResultsDf
```

### Setting different alpha values

```python
## First we will define an empty dataframe to store our values from the loop

resultsDf_6 = pd.DataFrame({'Alpha Values':[],'Train RMSE':[],'Train MAPE':[]
                                ,'Test RMSE': [],'Test MAPE': []})
resultsDf_6
```

```python
# Defining a loop which iterates from 0.1 to 1 at an interval of 0.1.
model_SES = SimpleExpSmoothing(SES_train['Rose'])
for i in np.arange(0.1,1,0.1):
    # Fitting the Simple Exponential Smoothing model with different alpha values got from the defined loop above
    model_SES_alpha_i = model_SES.fit(smoothing_level=i,optimized=False)
    # Storing the predicted values of the training set in the training data frame. 
    SES_train['predict',i] = model_SES_alpha_i.fittedvalues
    # Storing the predicted values of the test set in the test data frame
    SES_test['predict',i] = model_SES_alpha_i.forecast(steps=len(test))
    
    # Evaluating the predictions on the training set using the RMSE and MAPE
    rmse_model5_train_i = metrics.mean_squared_error(SES_train['Rose'],SES_train['predict',i],squared=False)
    mape_model5_train_i = calculate_mape(SES_train['Rose'],SES_train['predict',i])
    
    # Evaluating the predictions on the test set using the RMSE and MAPE
    rmse_model5_test_i = metrics.mean_squared_error(SES_test['Rose'],SES_test['predict',i],squared=False)
    mape_model5_test_i = calculate_mape(SES_test['Rose'],SES_test['predict',i])
    
    # Adding all the RMSE and MAPE values for both training and test set for their corresponding alpha values.
    resultsDf_6 = resultsDf_6.append({'Alpha Values':i,'Train RMSE':rmse_model5_train_i 
                                      ,'Train MAPE': mape_model5_train_i,'Test RMSE':rmse_model5_test_i
                                      ,'Test MAPE':mape_model5_test_i}, ignore_index=True)
```

### Setting different alpha and beta values for double exponential smoothing

```python
## First we will define an empty dataframe to store our values from the loop

resultsDf_alpha = pd.DataFrame({'Alpha Values':[],'Beta Values':[],'Train RMSE':[],'Train MAPE':[]
                                ,'Test RMSE': [],'Test MAPE': []})
resultsDf_alpha
```

```python
for i in np.arange(0.1,1.1,0.1): #Defining the alpha values from this loop
    for j in np.arange(0.1,1.1,0.1): #Defining the beta values from this loop
# We have defined a nested loop. This means one loop inside another. Here, due to this loop Python is going to fit the
# the Double Exponential model by varying different alpha and beta parameters. The way Python runs these set of loops is
# it fixes the value of 'i' (which is alpha) and varies all the values of 'j' (which is beta).

        # Fitting the Double Exponential Smoothing model with appropriate i and j values from the loop
        model_DES_alpha_i_j = model_DES.fit(smoothing_level=i,smoothing_slope=j,optimized=False)
        # Predicting on the training set
        DES_train['predict',i,j] = model_DES_alpha_i_j.fittedvalues
        # Predicting on the training set
        DES_test['predict',i,j] = model_DES_alpha_i_j.forecast(steps=len(DES_test))
        
        # Model Evaluation using RMSE and MAPE on for the training data predictions
        rmse_model6_train = metrics.mean_squared_error(DES_train['Rose'],DES_train['predict',i,j],squared=False)
        mape_model6_train = calculate_mape(DES_train['Rose'],DES_train['predict',i,j])
        
        # Model Evaluation using RMSE and MAPE on for the test data predictions
        rmse_model6_test = metrics.mean_squared_error(DES_test['Rose'],DES_test['predict',i,j],squared=False)
        mape_model6_test = calculate_mape(DES_test['Rose'],DES_test['predict',i,j])
        
        # Putting all the model evaluation results into one data frame with their corresponding alpha and beta values
        resultsDf_alpha = resultsDf_alpha.append({'Alpha Values':i,'Beta Values':j,'Train RMSE':rmse_model6_train
                                      ,'Train MAPE': mape_model6_train,'Test RMSE':rmse_model6_test
                                      ,'Test MAPE':mape_model6_test}, ignore_index=True)
```

### Setting different alpha, beta and gamma values for triple exponential smoothing

```python
resultsDf_beta = pd.DataFrame({'Alpha Values':[],'Beta Values':[],'Gamma Values':[],'Train RMSE':[],'Train MAPE':[]
                                ,'Test RMSE': [],'Test MAPE': []})
resultsDf_beta
```

```python
for i in np.arange(0.1,1.1,0.1): #Defining the alpha values from the loop
    for j in np.arange(0.1,1.1,0.1): #Defining the beta values from the loop
        for k in np.arange(0.1,1.1,0.1): #Defining the gamma values from the loop
# We have defined a nested loop. This means one loop inside another. Here, due to this loop Python is going to fit 
# the Triple Exponential model by varying different alpha and beta parameters. 


            model_TES_alpha_i_j_k = model_TES.fit(smoothing_level=i,smoothing_slope=j,smoothing_seasonal=k,optimized=False)
            TES_train['predict',i,j,k] = model_TES_alpha_i_j_k.fittedvalues
            TES_test['predict',i,j,k] = model_TES_alpha_i_j_k.forecast(steps=len(TES_test))
        
            rmse_model8_train = metrics.mean_squared_error(TES_train['Rose'],TES_train['predict',i,j,k],squared=False)
            mape_model8_train = calculate_mape(TES_train['Rose'],TES_train['predict',i,j,k])
        
            rmse_model8_test = metrics.mean_squared_error(TES_test['Rose'],TES_test['predict',i,j,k],squared=False)
            mape_model8_test = calculate_mape(TES_test['Rose'],TES_test['predict',i,j,k])
        
            resultsDf_beta = resultsDf_beta.append({'Alpha Values':i,'Beta Values':j,'Gamma Values':k,'Train RMSE':rmse_model8_train
                                      ,'Train MAPE': mape_model8_train,'Test RMSE':rmse_model8_test
                                      ,'Test MAPE':mape_model8_test}, ignore_index=True)
```
